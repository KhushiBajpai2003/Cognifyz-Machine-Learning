```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.preprocessing import LabelEncoder


# Handle missing values
df['Cuisines'] = df['Cuisines'].fillna('')  # Fill missing cuisines with an empty string

# Encode categorical variables using Label Encoding
label_encoder = LabelEncoder()
df['City'] = label_encoder.fit_transform(df['City'])

# Sample user preferences
user_preferences = {
    'City': 'Makati City',  # Add City to user preferences
    'Cuisines': 'Japanese',  # Add Cuisines to user preferences
    'Price Range': 1500
}

# Feature selection for content-based filtering
features = ['City', 'Cuisines']

# # Convert numerical values to string
# df['Average Cost for two'] = df['Average Cost for two'].astype(str)

# Concatenate selected features to create a text-based feature
df['features'] = df[features[1:]].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the data
tfidf_matrix = tfidf_vectorizer.fit_transform(df['features'])

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Function to get restaurant recommendations based on user preferences
def get_recommendations(user_preferences, cosine_sim, data):
    # Transform user preferences to match the features used in the model
    user_features = ' '.join([str(user_preferences[feature]) for feature in features[1:]])

    # Transform the user features using the same TF-IDF vectorizer
    user_tfidf = tfidf_vectorizer.transform([user_features])

    # Calculate the cosine similarity between user preferences and restaurants
    cosine_similarities = linear_kernel(user_tfidf, tfidf_matrix).flatten()

    # Get the indices of the top similar restaurants
    restaurant_indices = cosine_similarities.argsort()[::-1]

    # Return the recommended restaurants
    recommended_restaurants = data.iloc[restaurant_indices]

    return recommended_restaurants

# Get recommendations for the sample user preferences
recommended_restaurants = get_recommendations(user_preferences, cosine_sim, df)

# Display the recommended restaurants
print(recommended_restaurants[['City', 'Cuisines']])
```
